```python
rfm = pd.concat([R_score,F_score,M_score],axis= 1)
```

```python
def R_vs(x):
    if x > 3.82:
        return "高"
    else:
        return "低"
```

```python
rfm['R'] = rfm['R_score'].apply(lambda x : R_vs(x))
```

```python
rfm['R']
```

```python
def F_vs(x):
    if x > 2.02:
        return "高"
    else:
        return "低"
```

```python
rfm['F'] = rfm['F_score'].apply(lambda x : F_vs(x))
```

```python
def M_vs(x):
    if x > 2.02:
        return "高"
    else:
        return "低"
```

```python
rfm['M'] = rfm['M_score'].apply(lambda x : M_vs(x))
```

```python
rfm.head()
```

```python
rfm['rfm'] = rfm['R'] + rfm["F"] +rfm['M']
```

```python
rfm.head()
```

```python
def value(x):
    if x == "高高高":
        return "重要价值用户"
    elif x == "高低高":
        return "重要发展客户"
    elif x == "低高高":
        return "重要保持客户"
    elif x == "低低高":
        return "重要挽留客户"
    elif x == "高高低":
        return "一般价值用户"
    elif x == "高低低":
        return "一般发展客户"
    elif x == "低高低":
        return "一般保留客户"
    else:
        return "一般挽留客户"
```

```python
rfm['value'] = rfm['rfm'].apply(lambda x:value(x))
```
